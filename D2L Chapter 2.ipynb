#### 2.1.4. Broadcasting

2.1.3. 절에서 다루었던 Operation들은 모두 같은 shape끼리 연산을 수행하였다. 하지만 특정 조건을 만족한다면 shape가 달라도 연산이 가능하고, 이 메커니즘을 Broadcasting이라 한다.

```python
a = tf.reshape(tf.range(3), (3,1))
b = tf.reshape(tf.range(2), (1,2))

a, b
```

```python
a + b
```

위 코드에서 a는 $3 \times 1$, b는 $1 \times 2$ Tensor로 모양이 다르지만, $3 \times 2$ 모양의 Tensor가 결과로 나왔다.

#### 2.1.5. Saving Memory

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

연산을 실행하면 메모리 할당이 일어날 수 있다. Python에서는 Y+X를 수행한 결과를 새 메모리에 할당하고 Y가 그 메모리를 pointing하게끔 한다. 이는 메모리 낭비로 이어질 수 있다.

```python
Z = tf.Variable(tf.zeros_like(Y))
print('id(Z): ', id(Z))
Z.assign(X+Y)
print('id(Z): ', id(Z))
```

위와 같이 Variable 컨테이너를 선언하고 assign을 통해 Z 안에서 연산을 수행하는 방법 (in-place)으로 해결할 수 있다. zeros_like 메소드는 argument와 shape가 같고 0으로 채워진 Tensor를 반환한다.

```python
@tf.function
def computation(X, Y):
  Z = tf.zeros_like(Y)
  A = X + Y
  B = A + Y
  C = B + Y
  return C + Y

computation(X, Y)
```

위와 같이 @tf.function이라는 decorator를 붙여서 자동으로 최적화를 수행할 수 있다. 이 또한 메모리 낭비를 줄일 수 있는 방법 중 하나이다.

Tensor는 머신러닝에서 데이터 저장 및 조작을 위한 코어이다. Tensor는 위와 같이 다양한 Operation, Broadcasting과 더하여 메모리 낭비도 줄여주고 다른 Python 자료형으로도 변환이 가능한 편리한 클래스다, 라고 요약할 수 있다.

## 2.3. **Linear Algebra**

#### 2.3.1. Scalars

```python
x = tf.constant(3.0)
y = tf.constant(2.0)
z = tf.constant([3.0])
x + y, x * y, x / y, x**y, z
```

위와 같이 list 형태가 아닌, numeric value를 argument로 주면 scalar를 생성할 수 있다. scalar는 shape가 z와 다르게게 ()이다.

#### 2.3.2. Vectors

```python
x = tf.range(3)
y = tf.random.normal([4])
z = tf.constant([2,56,3.0])
x,y,z
```

위와 같이 다양한 방법으로 벡터를 생성할 수 있다. 스칼라는 0<sup>th</sup>-order Tensor이고, 벡터는 1<sup>st</sup>-order Tensor이다. 뒷 절에서 다루는 행렬은 2<sup>nd</sup>-order Tensor이다.

#### 2.3.3. Matrices

```python
A = tf.reshape(tf.range(6), (3, 2))
B = tf.random.normal([3,2])
C = tf.constant([[1,3,7],[2,7,2]])
A,B,C
```

벡터와 유사하게 다양한 방법으로 행렬을 생성할 수 있다. 위에서 언급했듯 행렬은 2<sup>nd</sup>-order Tensor이다.

```python
tf.transpose(A)
```

전치 연산은 transpose 메소드로 가능하다.

```python
A = tf.constant([[1, 2, 3], [2, 0, 4], [3, 4, 5]])
A == tf.transpose(A)
```

대칭행렬은 전치한 행렬과 같기 때문에 모든 원소가 True인 모습이다.

#### 2.3.4. Tensors

```python
A = tf.reshape(tf.range(24), (2, 3, 4))
B = tf.random.normal([3,2,4])
C = tf.constant([[[1,3,7],[2,7,2]],[[5,2,7],[1,2,9]]])
A,B,C
```

역시 다양한 방법으로 텐서를 생성할 수 있다. 스칼라도 텐서이고, 벡터도 텐서이고, 행렬도 텐서이다. 텐서는 저 셋을 모두 포괄하여, 더 나아가서 일반적인 형식으로 이해하였다. 스칼라든, 벡터든, 행렬이든, 텐서든 생성 방식이 다 똑같다는 점을 알 수 있다.

#### 2.3.5. Basic Properties of Tensor Arithmetic

```python
A = tf.reshape(tf.range(6, dtype=tf.float32), (2, 3))
B = A  # No cloning of A to B by allocating new memory
A, A + B
```

위는 (2,3) shape의 텐서를 생성하고, 같은 두 텐서를 더한 코드이다.

```python
A * B
```

위는 행렬곱과는 다른, Element-wise product를 수행하는 Hadamard Product이다. Hadamard Product는 $\odot$으로 표현한다.

```python
a = 2
X = tf.reshape(tf.range(24), (2, 3, 4))
a + X, (a * X)
```

Broadcasting을 적용하여 스칼라와 텐서를 각각 더하고 곱한 코드이다. 덧셈의 경우 X와 차원이 같고 모든 원소가 2인 텐서를 더한 것이고, 곱셈은 상수배를 한 것이다. 같은 차원이 결과로 나온 것으로 보아, 텐서도 벡터 공간과 같이 텐서 공간을 이룬다는 것으로 이해하였다. 이에 대해 좀 더 찾아보니, 텐서공간은 텐서덧셈과 텐서곱이 정의된 공간이라고 한다. 즉 저 코드에서 곱셈은 X와 차원이 같고 모든 원소가 2인 텐서를 Hadamard Product를 수행한 것이다.

2.1절에서 본 것과 같이 reduce_sum 메소드로 모든 원소의 합을 구할 수 있다. 결과는 스칼라로 나온다.

```python
A.shape, tf.reduce_sum(A)
```

Axis argument를 주어 축 별로 합을 계산할 수 있다.

위와 같이 모든 축에 대한 합을 구하면 그냥 행렬만 넘겨주는 것과 결과가 같음을 볼 수 있다.

```python
tf.reduce_mean(A), tf.reduce_sum(A) / tf.size(A).numpy()
```

```python
tf.reduce_mean(A, axis=1), tf.reduce_sum(A, axis=1) / A.shape[1]
```

reduce_mean 메소드로 reduce_sum과 유사하게 축별로 평균을 구할 수 있다.

#### 2.3.7. Non-Reduction Sum

```python
sum_A = tf.reduce_sum(A, axis=1, keepdims=True)
sum_A, sum_A.shape
```

keepdims argument를 이용하여 차원을 유지한 채로 합을 구할 수 있다. 물론 합을 구하면 해당하는 축의 차원은 1이 될 것이다.

```python
A / sum_A
```

sum_A를 이용하여 axis 1에 해당하는 각 원소의 합을 1로 만드는 코드이다. axis를 1로 설정하였으므로, 안쪽 리스트 당 그 원소의 합이 1임을 볼 수 있다.

```python
tf.cumsum(A, axis=0)
```

```python
tf.cumsum(A, axis=1)
```

위 코드는 cumsum 메소드로 축을 따라 누적합을 계산하는 코드이다. axis가 0일 때는 행을 따라 누적합을 구하였고, axis가 1일 때는 열을 따라 누적합을 구한 모습이다.

#### 2.4.2. Visualization Utilities

matplotlib 패키지를 통해 함수 개형과 접선 등등을 시각화할 수 있다.

(#@save 주석이 코랩에서 지원되지 않는 것 같다. #@save는 d2l 패키지에서 반복적인 코드 작성을 피하게끔 해주지만 여기서는 #@save 사용 없이 써야 할 것 같다.)

```python
def use_svg_display():
    """Use the svg format to display a plot in Jupyter."""
    backend_inline.set_matplotlib_formats('svg')

def set_figsize(figsize=(3.5, 2.5)):
    """Set the figure size for matplotlib."""
    use_svg_display()
    d2l.plt.rcParams['figure.figsize'] = figsize

def set_axes(axes, xlabel, ylabel, xlim, ylim, xscale, yscale, legend):
    """Set the axes for matplotlib."""
    axes.set_xlabel(xlabel), axes.set_ylabel(ylabel)
    axes.set_xscale(xscale), axes.set_yscale(yscale)
    axes.set_xlim(xlim),     axes.set_ylim(ylim)
    if legend:
        axes.legend(legend)
    axes.grid()

def plot(X, Y=None, xlabel=None, ylabel=None, legend=[], xlim=None,
         ylim=None, xscale='linear', yscale='linear',
         fmts=('-', 'm--', 'g-.', 'r:'), figsize=(3.5, 2.5), axes=None):
    """Plot data points."""

    def has_one_axis(X):  # True if X (tensor or list) has 1 axis
        return (hasattr(X, "ndim") and X.ndim == 1 or isinstance(X, list)
                and not hasattr(X[0], "__len__"))

    if has_one_axis(X): X = [X]
    if Y is None:
        X, Y = [[]] * len(X), X
    elif has_one_axis(Y):
        Y = [Y]
    if len(X) != len(Y):
        X = X * len(Y)

    set_figsize(figsize)
    if axes is None:
        axes = d2l.plt.gca()
    axes.cla()
    for x, y, fmt in zip(X, Y, fmts):
        axes.plot(x,y,fmt) if len(x) else axes.plot(y,fmt)
    set_axes(axes, xlabel, ylabel, xlim, ylim, xscale, yscale, legend)

x = np.arange(0, 3, 0.1)
plot(x, [f(x), 2 * x - 3], 'x', 'f(x)', legend=['f(x)', 'Tangent line (x=1)'])
```

## 2.5. **Automatic Differentiation**

#### 2.5.1. A Simple Function

```python
x = tf.range(4, dtype=tf.float32)
x
```

```python
x = tf.Variable(x)
x
```

```python
# Record all computations onto a tape
with tf.GradientTape() as t:
    y = 2 * tf.tensordot(x, x, axes=1)
y
```

```python
x_grad = t.gradient(y, x)
x_grad
```

```python
x_grad == 4 * x
```

위 코드들은 벡터 $x$에 대해 $y = 2x^{T}x$에 대한 Gradient를 관측한다. 당연한 사실이지만, 마지막 줄 코드로 Gradient는 4$x$로 표현될 수 있음을 알 수 있다. 

```python
with tf.GradientTape() as t:
    y = x * x
t.gradient(y, x)  # Same as y = tf.reduce_sum(x * x)
```

이 경우 Gradient는 2$x$가 되겠다.

#### 2.5.3 Detaching Computation

```python
# Set persistent=True to preserve the compute graph.
# This lets us run t.gradient more than once
with tf.GradientTape(persistent=True) as t:
    y = x * x
    u = tf.stop_gradient(y)
    z = u * x

x_grad = t.gradient(z, x)
x_grad == u
```

위 코드는 편미분과 언뜻 비슷해 보이지만, 약간 다른 기능에 대해 보여준다. 원래는 $y$가 $x$에 관한 식이기 때문에 $y$를 고정하고 $x$에 대한 $z$의 편미분을 구한다는 것은 이상해 보이지만, 위 코드는 실제로 그렇게 한다. $y$를 stop_gradient를 통해 $u$라는 어떤 고정값을 두고, 거기서 $x$ 에 대한 편미분을 구한다. 이 때 $z$는 $ux$로 나타내기 때문에 편미분은 $x$가 된다. d2l의 설명을 보면 "we can create a new variable u that takes the same value as y but whose provenance (how it was created) has been wiped out." 라고 설명한다. $y$가 어떤 변수에 관한 식인지에 대한한 그런 걸 싹 다 무시한다고 나와있다.

```python
t.gradient(y, x) == 2 * x
```

$y$에 대해서 여전히 정상적으로 $2x$로 나타나는 것을 볼 수 있다.

#### 2.5.4. Gradients and Python Control Flow

```python
def f(a):
    b = a * 2
    while tf.norm(b) < 1000:
        b = b * 2
    if tf.reduce_sum(b) > 0:
        c = b
    else:
        c = 100 * b
    return c

a = tf.Variable(tf.random.normal(shape=()))
with tf.GradientTape() as t:
    d = f(a)
d_grad = t.gradient(d, a)
d, a, d_grad
```

```python
d_grad == d / a
```

위 예시는 조건문이나 반복문 등 Python의 Control Flow가 포함된 함수에서도 Gradient를 구할 수 있다는 것을 보여주는 코드이다. f를 보면, 반복문과 조건문이 포함되어 있다. 언뜻 보면 Gradient를 구하는 것이 까다로워 보이지만, 결국 f는 a에 대한 선형식이기 때문에, f(a)/a가 Gradient가 된다. 코드에서 알 수 있듯이 f처럼 복잡한 함수에 대해 gradient 메소드로 Gradient를 간단히 구한 것과 같다. 딥러닝에서 Control Flow가 밥먹듯이 많이 쓰이기 때문에, 이러한 기능은 중요하다.

## 2.6. **Probability and Statistics**

```python
%matplotlib inline
import random
import tensorflow as tf
from tensorflow_probability import distributions as tfd
from d2l import tensorflow as d2l
```

#### 2.6.1. A Simple Example: Tossing Coins

```python
num_tosses = 100
heads = sum([random.random() > 0.5 for _ in range(100)])
tails = num_tosses - heads
print("heads, tails: ", [heads, tails])
```

확률의 단순한 예로 동전 던지기를 구현한 코드이다.

```python
fair_probs = tf.ones(2) / 2
tfd.Multinomial(100, fair_probs).sample()
```

```python
tfd.Multinomial(100, fair_probs).sample() / 100
```

다항분포를 이용한 좀 더 일반화된 코드이다.

```python
counts = tfd.Multinomial(10000, fair_probs).sample()
counts / 10000
```

표본이 많을수록 오차가 적음을 보여주는 코드이다. 오차가 ${1} \over \sqrt{n}$에 비례한다는 것은 잘 알려진 사실이다.

```python
counts = tfd.Multinomial(1, fair_probs).sample(10000)
cum_counts = tf.cumsum(counts, axis=0)
estimates = cum_counts / tf.reduce_sum(cum_counts, axis=1, keepdims=True)
estimates = estimates.numpy()

d2l.set_figsize((4.5, 3.5))
d2l.plt.plot(estimates[:, 0], label=("P(coin=heads)"))
d2l.plt.plot(estimates[:, 1], label=("P(coin=tails)"))
d2l.plt.axhline(y=0.5, color='black', linestyle='dashed')
d2l.plt.gca().set_xlabel('Samples')
d2l.plt.gca().set_ylabel('Estimated probability')
d2l.plt.legend();
```
